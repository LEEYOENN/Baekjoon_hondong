```python
def solution(friends, gifts):
    
    give = {name: 0 for name in friends}
    take = {name: 0 for name in friends}
        
    history = {giver: {receiver: 0 for receiver in friends} for giver in friends}
    
    for record in gifts:
        sender, receiver = record.split(" ")
        give[sender] += 1
        take[receiver] += 1
        history[sender][receiver] += 1
        
    gift_index = {}
    for name in friends:
        gift_index[name] = give[name] - take[name]
    
    next_gift = {name: 0 for name in friends}
    
    for giver in friends:
        for receiver in friends:
            if giver == receiver: continue
            
            if history[giver][receiver] > history[receiver][giver]:
                next_gift[giver] += 1
            elif history[giver][receiver] == history[receiver][giver]:
                if gift_index[giver] > gift_index[receiver]:
                    next_gift[giver] += 1
            
    
    max_gift = -(float('inf'))
    
    for k, v in next_gift.items():
        if max_gift < v:
            max_gift = v
    
    answer = max_gift
    return answer
```
